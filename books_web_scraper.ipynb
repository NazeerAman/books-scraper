```python

from bs4 import BeautifulSoup
import requests
```

```python
total_pages=50

book_details=[]

for page_number in range(1,total_pages+1):
    url = f"https://books.toscrape.com/catalogue/page-{page_number}.html"
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    ol=soup.find("ol")
    artilces=ol.find_all("article",class_="product_pod")

    for article in artilces:
        image=article.find("img")
        title=image.attrs["alt"]

        star=article.find("p")
        star=star["class"][1]

        price=article.find("p",class_="price_color").text
        price=float(price[2:])

        availability=article.find("p",class_="instock availability").text.strip()

        book_details.append([title,price,star,availability])
```

```python
print(f"Total books: {len(book_details)}")
```

```python
book_details
```

```python
import pandas as pd
```

```python
df=pd.DataFrame(data=book_details,columns=["Title","Price(£)","Ratings","Availability"])
```

```python
df
```

```python
rating_map = {
    "One": 1,
    "Two": 2,
    "Three": 3,
    "Four": 4,
    "Five": 5
}
```

```python
df["Ratings"] = df["Ratings"].map(rating_map)
```
